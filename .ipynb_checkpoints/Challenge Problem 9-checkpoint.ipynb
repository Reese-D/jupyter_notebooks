## Modeling the problem
This part was fairly simple, I just added an edge between each person and their potential. This fits the edge coloring well because no person can play twice on the same day, so each edge connected to any individual have to be different colors. Each color will represent a match played on some day, EG. red might be monday. This makes certain that nobody can play twice on monday as no vertex will touch any color twice.

```python
dol = {
    'a':['d','e','f','g'],
    'b':['d','e','f','g'],
    'c':['d','e','f','g'],
    'd':['b','c','f','g'],
    'e':['b','c','f','g'],
    'f':['b','c','d','e'],
    'g':['b','c','d','e'],
}
```

## Visual
Here's a quick visual of the graph before any coloring is done to it

```python
g = nx.from_dict_of_lists(dol)
nx.draw(g,with_labels=True)
plt.show()
```

## Expand the dictionary
Here I just use networkx to convert my dict of list to a graphic then back again, this adds all the missing two way edges in the graph in, this way if I specify a dictionary like this { 1:[2,3],4:[1] } the returned dictionary will automatically add 4 to 1's list and 1 to 4's list for me. This makes processing the dictionary a little easier

```python
dol_full = nx.to_dict_of_lists(g)
print(dol_full)
```

## Approaching the problem
The idea behind this function is to color from high nodes to low nodes and color each edge without any vertice touching two colors. Once this is done we can simply map a day of the week to a color.

Here I create a function that does the follow:

#### sorts the dictionary
by the node's degree. Highest degree vertices get colored first, this increases the probablity of making it minimal

#### loops
through every vertice in the dictionary, and every neighbor of every vertex

#### finds
all of the colors that have already been used on a given node / neighbor combo

#### ignores
any edge pairs that have already been visited

#### disjunctive union
takes the list of all colors, and only returns the ones that do not exist in the list of used colors (not a full disjunctive union)

#### appends
adds a touple of the two edges and the color connecting them to the edges list, this is used to check whether or not we've visited this combination in the 'ignores' clause, it also is used in the 'finds' clause to determine which colors have been used for any given node

#### draws
finally the graph edge that is found is rendered

#### extra comments
The return function was primarily used for debugging, it also gives a text representation of the graph. On another note the color_graph function only exists to make calling get_edge_list a little simpler so that the list of colors doesn't have to be specified for every single call.


```python
def get_edge_list(dol_full, colors, g, pos):
    edge_list = []
    #sorts
    for vertex in sorted(dol_full, key = lambda v: len(dol_full[v]), reverse=True):
        print(vertex)
        #loops
        for neighbor in dol_full[vertex]:
            #finds
            used_colors = [x[2] for x in edge_list if x[0] in [vertex,neighbor] or x[1] in [vertex, neighbor]]
            print(used_colors)
            #ignores
            if (neighbor, vertex) in map(lambda x: (x[0], x[1]),edge_list):
                continue
            #disjunctive union
            unused_colors = list(filter(lambda x: x not in used_colors, colors))
            #appends
            edge_list.append((vertex, neighbor, unused_colors[0]))
            used_colors.append(unused_colors[0])
            #draws
            nx.draw_networkx_edges(g, pos, edgelist = [(vertex,neighbor)], edge_color = unused_colors[0], width = 4)
    return edge_list
    
def color_graph(dol_full, g, pos):
    return get_edge_list(dol_full, ['red','blue','green','yellow','orange','purple'], g, pos)
    
```

```python
pos = nx.circular_layout(g)
nx.draw_networkx_nodes(g, pos)
nx.draw_networkx_labels(g, pos)
graph = color_graph(dol_full, g, pos)
print(graph)
plt.axis('off')
plt.show()
```

#### Map it
Now is the simple matter of converting edges and colors to days of the week and matches! Below I zip days and colors together to get a touple that looks something like ('Sunday','red') for example. After that list comprehension makes it easy to match up (day,color) with (player,opponent,color) by creating a new touple of (day,player,opponent) whenever the color in both touples are equivalent.

```python
colors = ['red','blue','green','yellow','orange','purple'] #copied from above
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
mapping = zip(days,colors)

[(x[0], y[0], y[1]) for x in mapping for y in graph if y[2] == x[1]]
```
