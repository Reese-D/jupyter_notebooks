# Challenge Problem 7

## Part 1 - Creating the graph

#### To create the graph I went through each individual lane (1 - 12) and added any other lane that could go simultaneously during the same light. So for example turn lanes can only go at the same time as the two lanes adjacent to them, as well as the turn lane on the opposing side of the light. In this example 3,4,9,10 were turn lanes, you can see they all have exactly 3 linked lanes, as each of the four directions are mirrors of eachother. The other two lanes in any given direction have 4 other lanes that can proceed at the same time as them, their two adjacent lanes, and all of the lanes on the opposing side (with the exception of the left turn lane). Following these rules the graph below was constructed.

```python
graph = {1:[2,3,11,12], 
         2:[1,3,11,12],
         3:[1,2,10],
         4:[5,6,9],
         5:[4,6,7,8],
         6:[4,5,7,8],
         7:[8,9,5,6],
         8:[7,9,5,6],
         9:[4,7,8],
         10:[3,11,12],
         11:[10,12,1,2],
         12:[11,10,1,2]
        }
```

## Part 2 - Solving the problem

#### Minimum subgraphs
Once the graph is created there will be large number of connected lanes, to determine the minimum number of light changes I decided that it would be necessary to find the minimum number of star subgraphs that would exist while still including every node from the original graph. 

#### why it would work
In this way when a light changes one of these subgraphs can be eliminated, as all of them would be able to proceed at the same time. In this sense we're actually picking the smallest number of single nodes that are not in anyway connected to eachother, and still being at most 1 away from any individual node in the graph. The star graph below represents what one subgraph would look like, the idea is to find the smallest number of these that contain every node in the set.

```python
nx.draw(nx.star_graph(5))
plot.show()
```

## Part 3 - Solving the problem

#### Straightforward
Finding the minimum number of star graphs in the set is actually a problem that's been already solved, much to my surprise there existed a function that finds the maximal independent set of a graph

```python
g = nx.from_dict_of_lists(graph)
```

```python
tmp = list(nx.algorithms.connected_component_subgraphs(g))
#nx.draw(tmp)
#plot.show()
for graph in tmp:
    nx.draw(graph)
    plot.show()
print(tmp)
```
